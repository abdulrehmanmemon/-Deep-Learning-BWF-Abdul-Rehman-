```python
#merging two datasets:
import pandas as pd

# Create two dataframes
df1 = pd.DataFrame({'id': [1, 2, 3, 4],
                    'name': ['John', 'Mary', 'Joe', 'Emma']})
df2 = pd.DataFrame({'id': [1, 2, 4, 5],
                    'age': [30, 25, 40, 35]})

# Merge the dataframes using 'id' column as the key
merged_df = pd.merge(df1, df2, on='id')
merged_df
```

```python
# Concatenate the dataframes
concat_df = pd.concat([df1, df2])
concat_df
```

```python
# Create a dataframe
df = pd.DataFrame({'id': [1, 2, 3, 4],
                   'variable': ['A', 'A', 'B', 'B'],
                   'value': [10, 20, 30, 40]})

# Reshape the dataframe from long to wide format
wide_df = df.pivot(index='id', columns='variable', values='value')
wide_df
```

```python
df = pd.DataFrame({'id': [1, 2, 3, 4],
                   'A': [10, 20, 30, 40],
                   'B': [50, 60, 70, 80]})

# Reshape the dataframe from wide to long format
long_df = pd.melt(df, id_vars='id', var_name='variable', value_name='value')
long_df
```

```python
df1 = pd.DataFrame({'id': [1, 2, 3, 4],
                    'name': ['John', 'Mary', 'Joe', 'Emma']})
df2 = pd.DataFrame({'id': [1, 2, 4, 5],
                    'age': [30, 25, 40, 35]})

# Combine the dataframes using an outer join
combined_df = pd.merge(df1, df2, on='id', how='outer')
combined_df
```

```python
df1 = pd.DataFrame({'id': [1, 2, 3, 4],
                    'name': ['John', 'Mary', 'Joe', 'Emma']})
df2 = pd.DataFrame({'id': [1, 2, 4, 5],
                    'age': [30, 25, 40, 35]})

# Combine the dataframes using an inner join
combined_df = pd.merge(df1, df2, on='id', how='inner')
combined_df
```
