```python
import re
def Gender_prop(x):
  m= re.findall('(\d*)M', x)[0]
  f=re.findall('_(\d*)F', x)[0]
  if m=='':
    m='0'
  if f=='':
    f=0
  return int(m),int(f)
```

```python
df[['Male_prop','Female_prop']]=df['GenderPop'].apply(lambda x: pd.Series(Gender_prop(x)))
```

The first visualization your boss wants you to make is a scatterplot that shows average income in a state vs proportion of women in that state.

```python
df=df.rename(columns={'Female_prop':'Female_Pop',
                      'Male_prop':'Male_Pop'})
df.head()
```

```python
def compute_female_pop(row):
    if row['Female_Pop'] == 0:
        return row['TotalPop'] - row['Male_Pop']
    else:
        return row['Female_Pop']

# Apply the function to the Female_Pop column
df['Female_Pop'] = df.apply(compute_female_pop, axis=1)
```

```python
df['Male_prop']=df['Male_Pop']/df['TotalPop']
df['Female_prop']=df['Female_Pop']/df['TotalPop']
df.head()
```

```python
in_wp=df.groupby(['State','Female_prop']).agg({'Income': 'mean'}).reset_index()
```

```python
in_wp
```

```python
# plt.scatter(in_wp['Income'],in_wp['Female_prop'])
sns.regplot(x='Income',y='Female_prop',data=in_wp)
plt.xlabel('Avg Income')
plt.ylabel('Female_prop')
plt.title('Avg Income VS Female Prop for each state')
plt.show()
```
