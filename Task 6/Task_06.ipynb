* Loop through the list of people who should take the poll. If they have already taken the poll, print a message thanking them for responding. If they have not yet taken the poll, print a message inviting them to take

```python
print('\nExercise 6.4')
glossary = {
    'variable': 'A named memory location used to store data that can be modified during program execution.',
    'function': 'A named set of instructions that can be reused throughout a program.',
    'conditional statement': 'A programming construct that executes different code blocks depending on whether a certain condition is true or false.',
    'loop': 'A programming construct that allows you to repeat a set of instructions a certain number of times or until a certain condition is met.',
    'list': 'A collection of items that can be stored in a single variable, where each item can be accessed by its position or index.'
}

for word, meaning in glossary.items():
    print(f"{word.title()}: {meaning}\n")
print('\nExercise 6.5')
rivers = {
    'nile': 'egypt',
    'amazon': 'brazil',
    'yangtze': 'china'
}

# Print a sentence about each river
for river, country in rivers.items():
    print(f"The {river.title()} runs through {country.title()}.\n")

# Print the name of each river in the dictionary
for river in rivers.keys():
    print(river.title())

# Print the name of each country in the dictionary
for country in rivers.values():
    print(country.title())
print('\nExercise 6.6')

favorite_languages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

# List of people who should take the poll
people = ['jen', 'sarah', 'edward', 'phil', 'john', 'jessica', 'david']

# Loop through the list of people and print a message about whether they've taken the poll or not
for person in people:
    if person in favorite_languages:
        print(f"Thank you for taking the poll, {person.title()}!")
    else:
        print(f"{person.title()}, what's your favorite programming language? Please take the poll!\n")
```

```python
print('\nExercise 6.7')
person1 = {
    'first_name': 'John',
    'last_name': 'Doe',
    'age': 30,
    'city': 'New York'
}

person2 = {
    'first_name': 'Alice',
    'last_name': 'Smith',
    'age': 25,
    'city': 'Los Angeles'
}

person3 = {
    'first_name': 'Bob',
    'last_name': 'Johnson',
    'age': 40,
    'city': 'Chicago'
}

people = [person1, person2, person3]

for person in people:
    print(f"First Name: {person['first_name']}")
    print(f"Last Name: {person['last_name']}")
    print(f"Age: {person['age']}")
    print(f"City: {person['city']}")
    print()

print('\nExercise 6.8')
pet1 = {
    'name': 'Fluffy',
    'animal': 'cat',
    'owner': 'John'
}

pet2 = {
    'name': 'Max',
    'animal': 'dog',
    'owner': 'Alice'
}

pet3 = {
    'name': 'Spike',
    'animal': 'bird',
    'owner': 'Bob'
}

pets = [pet1, pet2, pet3]

for pet in pets:
    print(f"Name: {pet['name']}")
    print(f"Animal: {pet['animal']}")
    print(f"Owner: {pet['owner']}")
    print()

print('\nExercise 6.9')
favorite_places = {
    'John': ['New York', 'Paris', 'Tokyo'],
    'Alice': ['Los Angeles', 'Rome'],
    'Bob': ['Chicago', 'Sydney', 'London']
}

for person, places in favorite_places.items():
    print(f"{person}'s favorite places are:")
    for place in places:
        print(f"- {place}")
    print()

print('\nExercise 6.10')
favorite_numbers = {
    'John': [5, 7, 9],
    'Alice': [2, 4],
    'Bob': [1, 3, 5]
}

for person, numbers in favorite_numbers.items():
    print(f"{person}'s favorite numbers are:")
    for number in numbers:
        print(f"- {number}")
    print()

print('\nExercise 6.11')
# Define a dictionary of cities with their respective information
cities = {
    'Tokyo': {'country': 'Japan', 'population': '13.5 million', 'fact': 'Tokyo is the most populous city in Japan and has been the capital since 1868.'},
    'New York City': {'country': 'United States', 'population': '8.3 million', 'fact': 'New York City is the largest city in the United States and is known for its skyscrapers and cultural landmarks.'},
    'Paris': {'country': 'France', 'population': '2.1 million', 'fact': 'Paris is known for its architecture, art, fashion, and cuisine, and is home to famous landmarks like the Eiffel Tower and the Louvre.'}
}

# Loop through the cities and print their information
for city, info in cities.items():
    print(f"{city} is in {info['country']}, has a population of {info['population']}, and is known for {info['fact']}.\n")

print('\nExercise 6.12')

rivers = {
    'nile': {'country': 'Egypt', 'length': '6,650 km', 'discharge': '2,830 m³/s'},
    'amazon': {'country': 'Brazil', 'length': '6,400 km', 'discharge': '209,000 m³/s'},
    'yangtze': {'country': 'China', 'length': '6,300 km', 'discharge': '31,900 m³/s'},
    'mississippi': {'country': 'United States', 'length': '3,780 km', 'discharge': '16,800 m³/s'},
    'volga': {'country': 'Russia', 'length': '3,692 km', 'discharge': '8,060 m³/s'}
}

# Print sentences about each river, including its discharge
for river, info in rivers.items():
    print(f"The {river.title()} runs through {info['country']} and is {info['length']} long. It has a discharge of {info['discharge']}.\n")
```
