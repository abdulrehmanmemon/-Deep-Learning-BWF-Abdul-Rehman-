```python
import numpy as np
import pandas as pd
```

```python
#creating dataframe
df=pd.DataFrame({'Name':["Ali","Hussain","Akbar","Abdul Rehman",np.nan,"Momina","Ali"],
                 "Age":[12,21,23,43,23,12,12],
                 "Height":[2.31,np.nan,5.64,7.89,np.nan,9.08,2.31]})
df.head()
```

```python
#lets drop the nan value in Name
df=df.dropna(subset=['Name'])
df.head()
```

```python
#reset the index
df=df.reset_index(drop=True)
df.head()
```

```python
df
```

```python
#impute nan value in height with mean 
df['Height'].fillna(df['Height'].mean(),inplace=True)
```

```python
df
```

```python
#observer first and last entry of dataframe they are duplicate lets drop duplicates
df=df.drop_duplicates()
df
```
